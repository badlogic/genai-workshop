We use scikit-learn to generate our dataset, as well as to apply k-means to it.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

We generate a dataset in 2D, made up of 3 "blobs" of points that are mostly well separated. We also plot it, coloring each point with the color of the "blob" it belongs to.

```python
# Generate a 2D dataset with 3 centers (clusters)
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Plot the generated dataset
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap="viridis")
plt.title('Generated 2D Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

Next we use k-means to identify 3 clusters in this dataset, and plot the result.

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=3)  # We choose 3 clusters because we generated the data with 3 centers
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the clustered dataset
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, label='Centroids')
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

Here's the problem though: generally, we do not know the number of clusters in a dataset. This is the `k` in k-means, also known as a hyper parameter. This is a common problem in real-world applications.

Let's cluster again, telling k-means to find 5 clusters.

```python
# Apply K-Means clustering
kmeans = KMeans(n_clusters=5)  # We choose 5 clusters this time
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the clustered dataset
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, label='Centroids')
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
